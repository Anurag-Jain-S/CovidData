```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('Covid.csv',index_col='State/UTs')
data
```

```python
data.info()
```

```python
plt.figure(figsize=(10,10))
data['Deaths'].sort_values(ascending=True)[:10].plot(kind='barh',color=['red','blue','green'])
plt.title('HIGHEST NUMBER OF DEATHS IN THE STATE/UT OF INDIA',fontsize=15)  #Note labelling the data
plt.xlabel('NO.OF DEATHS',fontsize=10)                 #Note labelling the x-label
plt.ylabel('STATE/UTs',fontsize=10)          #Note labelling the y-label
plt.show()
```

```python
plt.figure(figsize=(15,10))
data['Death Ratio'].sort_values(ascending=False)[:25].plot(kind='pie',autopct='%.1f %%',pctdistance=0.8)
plt.title('Highest Ratio Of Death in the State/UTs Of India',fontsize=15)   #Note labelling the data
plt.show()
```
